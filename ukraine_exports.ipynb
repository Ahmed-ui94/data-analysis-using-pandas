```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
ukrain_grain_exports = pd.read_excel("ukraine_grain_exports.xlsx")
```

```python
#Create a pivot table, 
#showing how many tons of each commodity (rows) have left each port (columns)
pd.pivot_table(data=ukrain_grain_exports, index="Commodity", 
columns="Departure port", values="Tonnage")
```

```python
#Create a pivot table, showing how many tons of each commodity (columns) 
# were going to each destination country (rows).
pd.pivot_table(data=ukrain_grain_exports, index="Country", columns="Commodity",
values="Tonnage")
```
