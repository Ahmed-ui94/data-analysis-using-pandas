```python
import pandas as pd 
import numpy as np 
import pyarrow as pyrr 
```

```python
df = (pd.read_csv("/Users/ahmedabukar/Downloads/pandas-workout-data/data/olympic_athlete_events.csv", 
                usecols=["Age", "Height", "Team", "Year", "Season", "Sport", "Medal"]))

df
```

```python
#only include games from 1980? 
df = df.loc[(df["Year"] >= 1980) & (df["Team"].isin(["Great Britain", "France", "United States", "Switzerland", "China", "Inida"]))]
```

```python
#what was the average age of olympic athletes? 
#in which country do players appear to consistently be youngest? ... China
df.pivot_table(index="Year", columns="Team", values="Age")
```

```python
#how tall were the tallest athletes in each year in each sport? 
df.pivot_table(index="Sport", columns="Year", values="Height", aggfunc="max")
```

```python
#how many medals did each country earn in each year? 
pd.pivot_table(df.dropna(subset="Medal"), index="Year", columns="Team", values="Medal", aggfunc="count")
```

```python
#create a pivot table that shows the number of medal that each team won per year
#with the index including the year and the season?
df.pivot_table(index=["Year", "Season"], columns="Team", values="Medal", aggfunc=np.size)
```

```python
#create pivot table that shows the both the average age and height per year per team?
df.pivot_table(index="Year", columns="Team", values=["Age", "Height"], aggfunc="max")
```

```python
#create a pivot table that shows the average age and the average height per year,
#per team, broken up by year and season
df.pivot_table(index=["Year", "Season"], columns="Team", values=["Age", "Height"], aggfunc="max")
```
