```python
columns_with_general = [
    "age",
    "are.you.datascientist",
    "is.python.main",
    "company.size",
    "country.live",
    "employment.status",
    "first.learn.about.main.ide",
    "how.often.use.main.ide",
    "main.purposes",
    "missing.features.main.ide",
    "nps.main.ide",
    "python.version.most",
    "python.years",
    "python2.version.most",
    "python3.version.most",
    "several.projects",
    "team.size",
    "use.python.most",
    "years.of.coding"
]
```

```python
def column_splitting(column_names):
    multi_column = []

    for column in column_names:
        if column not in columns_with_general:
            multi_column.append((column.rsplit(".", 1)[0], column.rsplit(".", 1)[1]))
        else:
            multi_column.append(("general", column))
    
    return multi_column
```

```python
py_df.columns = pd.MultiIndex.from_tuples(column_splitting(py_df.columns))
```

```python
py_df = py_df.sort_index(axis="columns")
```

```python
# remove rows from so_df in which language worked with is NaN?
so_df = so_df.dropna(subset="LanguageHaveWorkedWith")
```

```python
# remove rows in which python isn't included as a commonly used language?
so_df = so_df.loc[so_df["LanguageHaveWorkedWith"].str.contains(r"[Pp]ython", regex=True)]
```

```python
# Remove rows from so_df in which YearsCode in NaN? How many rows remain? ---38997
so_df = so_df.dropna(subset="YearsCode")
```

```python
# Replace the string value "Less than 1 year" in YearsCode?
so_df.loc[so_df["YearsCode"] == "Less than 1 year", "YearsCode"] = 0
```

```python
# Replace the string value "More than 50 years" in YearsCode?
so_df.loc[so_df["YearsCode"] == "More than 50 years", "YearsCode"] = 51
```

```python
# turn YearsCode column into an integer?
so_df["YearsCode"] = so_df["YearsCode"].astype(np.int16)
```

A create a new column in so_df called "experience", 
which will categorize the values in the 'YearsCode' into:
1. Less than 1 year
2. 1-2 years
3. 3-5 years
4. 6-10 years
5. 11+ years 

```python
so_df["experience"] = (pd
 .cut(so_df["YearsCode"], 
      bins=[so_df["YearsCode"].min(),
           2, 5, 10, 11, so_df["YearsCode"].max()],
     labels=["Less than 1 year", 
            "1-2 years",
            "3-5 years",
            "6-10 years",
            "11+ years"])
)
```
