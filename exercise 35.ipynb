```python
import pandas as pd
import numpy as np
```

```python
oecd_df = pd.read_csv("/Users/ahmedabukar/Downloads/:/data/oecd_locations.csv", 
                      header=None,
                     names=["country_abbrev", "country"],
                     index_col="country_abbrev")
oecd_df.head()
```

```python
oecd_tourism_df = (pd.read_csv("/Users/ahmedabukar/Downloads/:/data/oecd_tourism.csv",
                             usecols=["LOCATION", "TIME", "SUBJECT", "Value"],
                             index_col="LOCATION")
                   .loc[lambda df_ : df_["SUBJECT"] == "INT-EXP"]
                   .drop(columns="SUBJECT")
                  )
oecd_tourism_df
```

```python
# create a new series, tourism_speding in which the index reflects the country 
# names (i.e. not abbreviations) and the value contains the average tourism spending for that country
tourism_spending = pd.Series(oecd_tourism_df.groupby(oecd_df["country"])["Value"].mean())
tourism_spending.head()
```

```python
wine_df = pd.read_csv("/Users/ahmedabukar/Downloads/:/data/winemag-150k-reviews.csv",
                     usecols=["country", "points"])
wine_df.head()
```

```python
#Get the mean score for each country, across all wine reviews, sorted in descending order?
mean_wine_score = wine_df.groupby("country")["points"].mean().sort_values(ascending=False)
mean_wine_score.head()
```

```python
#perform a standard join between the average wine scores per country and the average tourism spending
# per country. Where do you see NaN values? What do that NaN values mean?
mean_wine_score.to_frame().join(tourism_spending)
```

```python
#find the correlation between average wine score and average tourism spending.
# what can you say about these two values? is there any correlation?
(mean_wine_score.to_frame().join(tourism_spending, how="outer")
 .corr()
)
```

```python
# join using outer join 
mean_wine_score.to_frame().join(tourism_spending, how="outer").head()
```
