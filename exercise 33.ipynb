```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/ahmedabukar/Downloads/pandas-workout-data/data/sat-scores.csv",
                usecols=["Year", "State.Code", "Total.Math", 
                         "Family Income.Less than 20k.Math",
                        "Family Income.Between 20-40k.Math",
                        "Family Income.Between 40-60k.Math",
                        "Family Income.Between 60-80k.Math",
                        "Family Income.Between 80-100k.Math",
                        "Family Income.More than 100k.Math"])
df
```

```python
df = df.rename({"Family Income.Less than 20k.Math": "Income<20k",
          "Family Income.Between 20-40k.Math" : "20k<Income<40k",
          "Family Income.Between 40-60k.Math" : "40k<Income<60k",
          "Family Income.Between 60-80k.Math" : "60k<Income<80k",
          "Family Income.Between 80-100k.Math" : "80k<Income<100k",
          "Family Income.More than 100k.Math" : "Income>100k"}, axis="columns")
df
```

```python
#find the average SAT math score for each income level, grouped and sorted by year
df.groupby("Year")[["20k<Income<40k", "40k<Income<60k",	"60k<Income<80k", 
                    "80k<Income<100k", "Income<20k", "Income>100k"]].expanding().mean().sort_index()
```

```python
#for each year in the data set, determine how much better each income group did, on average,
# then the next poorer group of students 
(df.groupby("Year")[["20k<Income<40k", "40k<Income<60k",	"60k<Income<80k", 
                    "80k<Income<100k", "Income<20k", "Income>100k"]]
                     .mean().T.pct_change())
```

```python
#which income levels consistently (i.e., across all years) do worse than the next-poorest group?
change = (df.groupby("Year")[["20k<Income<40k", "40k<Income<60k",	"60k<Income<80k", 
                    "80k<Income<100k", "Income<20k", "Income>100k"]]
                     .mean().T.pct_change().mean(axis="columns"))
change
```

```python
#calculate the descriptive statistics for all the changes in income brackets.
#where do you see the largest difference between income brackets?
change.dropna().describe()
```

```python
# which five states have the greatest gap in SAT math scores between the richest and poorest students/
df["gap_difference"] = df["Income>100k"] - df["Income<20k"]
df.groupby("State.Code")["gap_difference"].mean().sort_values(ascending=False).head()
```
