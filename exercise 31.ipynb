```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/ahmedabukar/Downloads/pandas-workout-data/data/oecd_tourism.csv",
                usecols=["LOCATION", "SUBJECT", "TIME", "Value"])
df
```

```python
df = df.replace("INT-EXP", "INT_EXP")
```

```python
#find the five countries that received the greatest amount of tourist dollars
#on average, across years in data set
df.loc[df["SUBJECT"] == "INT_REC"].groupby(["LOCATION", "TIME"])["Value"].mean().sort_values().tail()
```

```python
#find the five countries whose citizens spent the least amount of tourist dollars,
#on average, across years in the data set
df.loc[df["SUBJECT"] == "INT_REC"].groupby(["LOCATION", "TIME"])["Value"].mean().sort_values()
```

```python
locations_df = pd.read_csv("/Users/ahmedabukar/Downloads/pandas-workout-data/data/oecd_locations.csv",
                          names=["LOCATION", "country_name"], index_col="LOCATION")
locations_df
```

```python
#set the index of df to LOCATION?
df = df.set_index("LOCATION")
```

```python
#join the data frames?
joined_df = locations_df.join(df)
```

```python
joined_df
```

```python
#find the five countries whose citizens spent the least amount of tourist dollars,
#on average, across years in the data set
(joined_df.loc[joined_df["SUBJECT"] == "INT_REC"]
 .groupby(["LOCATION", "TIME"])
 ["Value"].mean()
 .sort_values()
 .head())
```

```python
#find the five countries that received the greatest amount of tourist dollars
#on average, across years in data set
(joined_df
 .loc[joined_df["SUBJECT"] == "INT_REC"]
 .groupby(["LOCATION", "TIME"])
 ["Value"].mean()
 .sort_values().tail())
```
