```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("/Users/ahmedabukar/Downloads/:/data/cities.json")
df.head()
```

```python
# create a bar plot showing how many of 1000 cities are in each state
# there should be one vertical bar per state
# the plot should ordered from smallest(left) to largest(right)
df.groupby("state")["city"].count().sort_values().plot.bar(figsize=(10, 10))
```

```python
df["growth_from_2000_to_2013"] = pd.to_numeric(df["growth_from_2000_to_2013"].str.replace(r"%", "", regex=True), errors="coerce")
```

```python
# create a bar plot comparing the growth of all cities in the state pennsylvania
(
    df
    .loc[df["state"] == "Pennsylvania"]
    .set_index("city")
    ["growth_from_2000_to_2013"]
    .sort_values()
    .plot.bar(figsize=(10, 10))
)
```

```python
# create a pie plot showing how much each Massachusetts city in the list contributes 
# to the overall population?
(
    df
    .loc[df["state"] == "Massachusetts"]
    .set_index("city")
    ["population"]
    .plot.pie(figsize=(15, 15))
)
```

```python
# create a scatter plot of the cities, putting the longitude on the x axis and latitude on the y axis
df.plot.scatter(x="longitude", y="latitude")
```

```python
# create a histogram of the growth rates among cities in both Texas and Michigan?
(
    df
    .loc[df["state"].isin(["Texas", "Michigan"])]
    .set_index("city")
    ["growth_from_2000_to_2013"]
    .plot.hist()
)
```

```python
# create a histogram of the growth rates among cities in both Taxas and California?
(
    df
    .loc[df["state"].isin(["Texas", "California"])]
    .set_index("city")
    ["growth_from_2000_to_2013"]
    .plot.hist()
)
```

```python
# create a bar plot from the average growth rate per state?
(
    df
    .groupby("state")
    ["growth_from_2000_to_2013"]
    .mean()
    .sort_values()
    .plot.bar(figsize=(12, 12))
)
```
